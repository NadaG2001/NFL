### predict the pass_result using NFL 2023 Big Data


```python
#import the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, mean_absolute_error
from sklearn.impute import SimpleImputer
```

```python
#function to convert gameClock time
def convertTimeToNumerical(t):
    t = t.split(":")
    if len(t) == 2:
        return int(t[0]) * 60 + int(t[1])
    return int(t[0]) * 3600 + int(t[1]) * 60 + int(t[2])
#function to convert players height
def heightInches(h):
    temp = list(map(int, h.split('-')))
    return temp[0]*12 + temp[1]
```

```python
# creating dataframe
#read csv files
dfPlays = pd.read_csv('../input/nfl-big-data-bowl-2023/plays.csv')
dfPlayer = pd.read_csv('../input/nfl-big-data-bowl-2023/players.csv')
dfWeek = pd.read_csv('../input/nfl-big-data-bowl-2023/week1.csv')
#converting values
dfPlays = pd.merge(dfWeek, dfPlays, on=['playId'])
dfPlays['gameClock'] = dfPlays['gameClock'].apply(convertTimeToNumerical)
dfPlays['height'] = dfPlayer['height'].apply(heightInches)
dfPlays = dfPlays.fillna(method='ffill')
# Selecting train and test data
xInput = dfPlays[['defendersInBox','quarter','yardsToGo','gameClock','height','s']]
yInput = dfPlays['passResult']
yInput.dropna()
```

```python
#split data into train and test
#week1
xTrain,xTest,yTrain,yTest = train_test_split(xInput,yInput,random_state=0,test_size=0.25)
#scaling dataset
scaler = MinMaxScaler()
xTrainScaled = scaler.fit_transform(xTrain)
xTestScaled = scaler.fit_transform(xTest)
```

### 1. DecisionTree Model

```python
#select model
model = DecisionTreeClassifier()
#train model
model.fit(xTrainScaled,yTrain)
predictions = model.predict(xTestScaled)
#observing performance metrics
accuracy = accuracy_score(yTest,predictions)*100
print("Accuracy of the model is {:.2f}%".format(accuracy))
```

```python
## saving Decision Tree Model
import pickle

# Save the model to a file
with open('DecisionTree_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
